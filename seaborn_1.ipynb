```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

### Seaborn Funtions

* Figure level Fucntions
* Axis level Functions

```python
tip = sns.load_dataset('tips')
tip.tail()
```

### Seaborn Classification

* Relational plot -> scatterplot, line plot
* Distribution plot -> Histogram,KDE plot, Rugplot
* Categorical plot -> Barplot, countplot,Boxplot,violinplot,swarmplot
* Matrix -> Heatmap, Clustermap
* Regression -> Linear,Logistic
* Multi Line Plot -> Pairplot, Jointplot

```python
## axis level function in seaborn
sns.scatterplot(data=tip,
                x='total_bill',
                y='tip',
                color='purple',
                s=tip['size']*50,
                edgecolor='blue',
                hue='sex',
                style='smoker')
```

```python
## figure level function in seaborn

sns.relplot(data=tip,
            x='total_bill',
            y='tip',
            kind='scatter',
            s=tip['size']*80,
            edgecolor='blue',
            hue='sex',
            palette='viridis')
```

### Distribution Plot

* Univariate analysis
* central Tendency
* Range of observations
* used to find out distribution

#### Types Of plot

* Histplot
* KDEplot
* Rugplot

```python
import plotly.express as px
iris = px.data.iris()
iris.head()
```

```python
## ploting Histogram figurelevel funtion
sns.displot(kind='hist',data=iris,x='sepal_length',hue='species',element='step')
```

```python
# ploting histogram axis level function
sns.histplot(data=iris,x='sepal_length',hue='species')
```

```python
## facet graphs
sns.displot(kind='hist',data=iris,x='sepal_length',col='species',hue='species_id',element='step',palette='viridis')
```

```python
### ploting KDE plot
tip.head()
```

```python
## axis level functions

sns.kdeplot(data=tip,x='total_bill',hue='sex')
```

```python
## fill -> in KDE plot
sns.kdeplot(data=tip,x='total_bill',hue='sex',fill=True)
```

```python
## figure level fucntion
sns.displot(kind='kde',data=tip,x='tip',hue='sex',fill=True)
```

```python
## Faceting using displot
sns.displot(kind='kde',data=tip,x='tip',col='sex',fill=True,hue='time',edgecolor='orange')
```

```python
## Rugplot
### plots marginal distribution by drawing tick along x and y 
plt.style.use('ggplot')
plt.figure(figsize=(10,10))
sns.kdeplot(data=tip,x='tip',fill=True,hue='time',edgecolor='orange')
sns.rugplot(data=tip,x='tip',hue='sex')
```

```python
## bivariate histogram
### histogram bins the data within rectangle that tile the plot and then shows count of observation with an each rectangle with fill color

sns.histplot(data=tip,x='total_bill',y='tip',hue='sex')
```

```python
## bivariate kdeplot
### bivariate kdeplot smoothes the(x,y) observations within 2d gaussian

sns.kdeplot(data=tip,x='total_bill',y='tip')
```

```python
Asia = gap_df[gap_df['continent'] == 'Asia']
Asia = Asia.pivot(index='country',values='lifeExp',columns='year')
Asia.head(10)
```

```python
## Axes level function it only works at axes level function
plt.figure(figsize=(10,10))
sns.heatmap(data=Asia,linewidths=0.5,annot=True)
```

```python
## European countries gdppercapita

Europe = gap_df[gap_df['continent'] == 'Europe'].pivot_table(index='country',columns='year',values='gdpPercap',aggfunc='mean')
Europe.head(10)
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data=Europe,cmap='winter_r',linewidths=0.5,annot=True)
```

```python
### Americas population

America = gap_df[gap_df['continent'] == 'Americas'].pivot(index='country',columns='year',values='pop')
America.head(10)
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data=America,linewidths=0.5,cmap='Blues_r',annot=True)
```
