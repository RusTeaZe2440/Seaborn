```python
import pandas as pd
import seaborn as sns
import numpy as  np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tip = sns.load_dataset('tips')
iris = px.data.iris()
```

```python
iris.head()
```

```python
tip
```

### Categorical Scatter plot

* Strip plot
* Swarm Plot

### Strip Plot

```python
## Strip plot -> numerical vs categorical data

sns.catplot(data=tip,x='day',y='tip',kind='strip')
```

### Swarm Plot

```python
## Jitter fucntion -> removes unnecesary noise from graph 

sns.catplot(data=tip,x='day',y='tip',kind='strip',jitter=False)
```

```python
plt.style.use('fivethirtyeight')
sns.catplot(data=iris,x='species',y='sepal_length',kind='swarm')
```

### Categorical Distribution Plot 

* Box Plot
* violin Plot

### Box Plot

A box plot is a standardized way of displaying the distribution of data based on five number summary
* Minimum
* First Quartile[Q1]
* Median 
* Third Quartile[Q3]
* Maximum


Also helps to identify outliers

```python
plt.figure(figsize=(3,3))
plt.style.use('default')
sns.boxplot(data=iris,x='species',y='sepal_length')
```

```python
# figure Level fucntion
plt.style.use('default')
sns.catplot(kind='box',data=tip,x='day',y='total_bill',hue='sex')
```

### violin Plot

```python
### violin plot 
# axes level functioning

plt.figure(figsize=(5,3))
sns.violinplot(data=iris,x='species',y='petal_length')
```

### Categorical Estimat plot -> central tendency

* Bar plot
* Point plot
* Countplot

### Barplot

```python
### Bar plot

plt.style.use('default')
sns.catplot(data=tip,x='sex',y='total_bill',kind='bar')
```

```python
## estimator is used to calculate aggregate functions of bar plot
sns.barplot(data=tip,x='day',y='total_bill',hue='sex',estimator=np.max)
```

### Pointplot

```python
## Pointplot 
sns.pointplot(data=tip,x='day',y='total_bill',hue='sex',estimator=np.max)
```

```python
sns.catplot(kind='point',data=tip,x='sex',y='total_bill',estimator=np.mean)
```

### Countplot

```python
## countplot
sns.catplot(kind='count',data=tip,x='sex')
```

```python
## axis level functioning
sns.countplot(data=tip,x='sex',hue='day')
```

```python
sns.regplot(data=tip,x='total_bill',y='tip')
```

```python
## we can add hue parameter in lmplot

sns.lmplot(data=tip,x='total_bill',y='tip',hue='sex')
```
