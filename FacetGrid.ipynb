```python
import seaborn as sns
import plotly.express as px
```

```python
iris = px.data.iris()
tips = sns.load_dataset('tips')
```

### Facet Grid

```python
# traditional way of creting facet graphs

sns.relplot(data=tips,x='sex',y='total_bill',col='day',kind='scatter',row='time')
```

```python
sns.catplot(kind='bar',data=tips,x='sex',y='total_bill',ci=None,col='day',row='time')
```

```python
iris.head()
```

```python
sns.displot(kind='kde',data=iris,x='sepal_length',fill=True,hue='species',col='species')
```

### PairPlot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(data=tips)
g.map(sns.histplot)
```
